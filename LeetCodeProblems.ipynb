#1 Goal Parser Interpretation

You own a goal parser that can interpret a string command. The command consists of an alphabet "G", "()" and/or "(al) in some order. The goal parser will interpret  "G" as the string "G", "()" as the string "O" and "(al)" as the stroing "al". The interpreted strings are then concatenated in the orginal order. Given the string command, return the Goal Parser's interpretation of the command

```python
class Solution:
    def interpret(self, command:str) -> str:
        return command.replace("()","o").replace("(al", "al")
```

#2 Destination City

You are given the array paths, where path[i] = [cityAi, cityBi] means there exists a direct path going from cityAi to cityBi. Return the destination city, i.e, the city w/o any path going to another city. It's guaranteed that the graph of paths forms a line w/o any loop. Therefore, there will be exactly one destination city

```python
class Solution:
    def destCity(self, paths):
        outgoingCount = {}
        for path in paths:
            city_a, city_b = path
            outgoingCount[city_a] = outgoingCount.get(city_a, 0) + 1
            outgoingCount[city_b] = outgoingCount.get(city_b, 0) 
        
        for city in outgoingCount:
            if outgoingCount[city] == 0:
                return city

#alt sol

class Solution:
    def destCity(self, paths) -> str:
        A, B = map(set, zip(*paths))
        return (B-A).pop()
```

#3 How many numbers are smaller than the current number

Given the array nums, for each nums[i] find out how many numbers in the array are smaller than it. i.e, for each nums[i] you have to count the number of valid j's such that j!=i and nums[j]< nums[i]
return the answer in an array

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums):
        output = []

        for i in range(len(nums)):
            count = 0
            for k in nums:
                if k != i and  nums[k] < nums[i]:
                    count += 1
            output.append(count)
        return output   

#alt sol
class Solution:
    def smallerNumbersThanCurrent(self, nums):
        sortedNums = sorted(nums, reverse = True) #descending order
        smallerCount = {}
        for i in range(len(sortedNums)-1):
            num = sortedNums[i]
            currNum = sortedNums[i]
            nextNum = sortedNums[i+1]
            if nextNum < currNum:
                remainingValues = len(sortedNums)-(i + 1)
                smallerCount[currNum] = remainingValues
        smallerCount[sortedNums[-1]] = 0

        output = []
        for num in nums:
            output.append(smallerCount[num])
        
        return output

         
```
